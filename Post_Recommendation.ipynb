```python
#reading the file and hiding index given by inbuilt function
post.to_excel('Task - 1.xlsx',sheet_name='Raw_data',index=False)
```

```python
#Finding the number of unique users
users_n = post.User_id.unique().shape[0]
print(users_n)
```

```python
#Finding the number of unique posts
posts_n = post.Post_id.unique().shape[0]
print(posts_n)
```

```python
#Fully describing each variable
post.describe()
```

```python
#defining new rating
score = ((.38)*post['Upvote']*(1/190)) - ((.14)*post['Downvote']*(1/175)) + ((.50)*post['Views']*(1/200)) + ((.21)*post['Comment']*(1/50)) + ((.16)*post['Share']*(1/40)) - ((.15)*post['Time Spent']*(1/60)) + ((.04)*post['Link Opened']*(1/2))
```

```python
#Making a new column according to the calculated new scores
post["Score"] = score
post.head()
```

```python
#view of the first 30 rows of the sheet along with score
post.head(30)
```

```python
#data_matrix is the user x post matrix, which has rating of user i for the post j

data_matrix = np.zeros((users_n+1, posts_n+1))
```

```python
# data matrix . rows are users and columns are variables. the value at data_matrix[i][j]=rating for post j by user i.
for line in post.itertuples():
    data_matrix[line[1],line[2]]=line[10]
```

```python
#Similarities are found using user-user similarity and cosine matrix
from sklearn.metrics.pairwise import pairwise_distances 
user_similarity = pairwise_distances(data_matrix, metric='cosine')
post_similarity = pairwise_distances(data_matrix.T, metric='cosine')
```

```python
#A predict function is defined that will be used to define
def predict(ratings, similarity):
    pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
#prediction is done using predict function
post_prediction = predict(data_matrix, post_similarity)
```

```python
#Getting the index of the predicted posts in ascending order
post_prediction.argsort()
```

```python
#no_post= number of post want to suggest to each user
no_post=40
predict_post=np.argsort(-post_prediction)[:,:no_post] #making a new array with predictions of each users
predict_post[5] #visualization of prediction of a random user
```

```python
#giving user_ids to prediction for better viewing the results
predict_item=pd.DataFrame({'user_id':[] , 'post_ids':[]})

for i in range(1,users_n+1):
    predict_item = predict_item.append({'user_id': i, 'post_ids':predict_post[i]}, ignore_index=True)
    
predict_item.user_id.astype='int'
```

```python
#Viewing the prediction with user_ids
predict_item.head()
```

```python
#writing the prediction in excel sheet
writer = pd.ExcelWriter('Task - 1.xlsx', engine = 'xlsxwriter')
post.to_excel(writer, sheet_name = 'Raw_data', index = False)
predict_item.to_excel(writer, sheet_name = 'Predicted_output', index = False)
writer.save()
writer.close()
```
